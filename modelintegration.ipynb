# Stream and Batch Processing

```python
model.stream("write me a poem about love")
```

```python
for chunk in model.stream("write me a 200 words on artificial intelligence"):
    print(chunk.text, end="|", flush=True)
```

Batch - Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as it allows the model to process multiple requests in a single call.

```python
responses = model.batch(["why do parrots make good pets?",
"why do ants like to eat sugar?",
"what is quantum physics?"],
config = {"max_tokens":100,"max_concurrency": 5 #limit to 5 parallel requests}
)

for response in responses:
    print(response.content)
```
